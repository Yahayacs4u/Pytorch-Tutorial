# Building Linear R model

```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/master/delaney_solubility_with_descriptors.csv')
```

```python
df
```

```python
y = df['logS']
y
```

```python
x = df.drop('logS', axis=1)
x
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test, y_train,y_test = train_test_split(x,y,test_size=0.2, random_state=100)
```

```python
x_train
```

```python
x_test
```

# LR Model Training

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
y_lr_train_pred = lr.predict(x_train)
y_lr_test_pred = lr.predict(x_test)
```

```python
y_lr_train_pred
```

```python
y_lr_test_pred
```

```python
y_train
```

```python
y_lr_train_pred
```

# LR Model performance

```python
from sklearn.metrics import mean_squared_error, r2_score

lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)
```

italicized text

```python
print('LR MSE (Train): ', lr_train_mse)
print('LR R2 (Train): ', lr_train_r2)
print('LR MSE (Test): ', lr_test_mse)
print('LR R2 (Test): ', lr_test_r2)
```

```python
lr_result = pd.DataFrame(['Linear Regression',lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
lr_result.columns = ['Method','Training MSE','Training R2','Test MSE','Test R2']
```

```python
lr_result
```

RF Model Training

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(max_depth=2, random_state=100)
rf.fit(x_train, y_train)
```

# Applying model to make prediction

```python
y_rf_train_pred = rf.predict(x_train)
y_rf_test_pred = rf.predict(x_test)
```

# RF Model performance

```python
rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)
```

```python
rf_result = pd.DataFrame(['Random Forest',lr_train_mse, lr_train_r2, rf_test_mse, rf_test_r2]).transpose()
rf_result.columns = ['Method','Training MSE','Training R2','Test MSE','Test R2']
rf_result
```

# Models Comparison, LR and RF

```python
df_models = pd.concat([lr_result, rf_result], axis=0).reset_index(drop=True)
df_models
```

# OneHotEncoding using scikit learn

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

# Create the DataFrame
data = {
    'Fruit': ['Apple', 'Banana', 'Grape', 'Orange'],
    'Color': ['Red', 'Yellow', 'Green', 'Orange'],
    'Size': ['Small', 'Medium', 'Small', 'Large']
}
df = pd.DataFrame(data)

# Initialize the encoder
encoder = OneHotEncoder(drop='first', sparse_output=False)

# Fit and transform the data
encoded_data = encoder.fit_transform(df[['Color', 'Size']])

# Create a DataFrame for encoded features
encoded_df = pd.DataFrame(
    encoded_data,
    columns=encoder.get_feature_names_out(['Color', 'Size'])
)

# Combine with original non-categorical columns
final_df = pd.concat([df['Fruit'], encoded_df], axis=1)
print(final_df)
```

# OneHotEncoding using pandas

```python
import pandas as pd

# Create the DataFrame
data = {
    'Fruit': ['Apple', 'Banana', 'Grape', 'Orange'],
    'Color': ['Red', 'Yellow', 'Green', 'Orange'],
    'Size': ['Small', 'Medium', 'Small', 'Large']
}
df = pd.DataFrame(data)

# Perform one-hot encoding
df_encoded = pd.get_dummies(df, columns=['Color', 'Size'], drop_first=True)
print(df_encoded)
```
